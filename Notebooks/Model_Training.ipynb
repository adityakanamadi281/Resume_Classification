```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
resume_data=pd.read_csv(r"C:\Users\adity\Resume_Classification\Resumes\Cleaned_Resumes.csv")
resume_data
```

```python
x = resume_data['Resume_Details'].values
y = resume_data['Category'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=45, test_size=0.25)
x_train.shape, x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vector = TfidfVectorizer(sublinear_tf=True,stop_words='english')
```

```python
x_train = tfidf_vector.fit_transform(x_train)
x_test = tfidf_vector.transform(x_test)
```

```python
x_train.shape, x_test.shape
```

```python
print(y_train),print(y_test)
```

## Decision Tree Clasdsifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_DT = DecisionTreeClassifier(criterion='gini')
```

```python
model_DT.fit(x_train, y_train)
```

```python
y_pred = model_DT.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, classification_report, precision_score, recall_score
```

```python
accuracy_DT = accuracy_score(y_test, y_pred)
```

```python
print('Accuracy of training set : {:.2f}'.format(model_DT.score(x_train, y_train)))
print('Accuracy of  test set    : {:.2f}'.format(model_DT.score(x_test, y_test)))
```

```python
precision_DT = round(precision_score(y_test,y_pred,average = 'macro'),2)
recall_DT= round(recall_score(y_test,y_pred, average = 'macro'),2)
f1_DT = round(f1_score(y_test,y_pred, average = 'macro'),2)
accuracy_DT = round(accuracy_score(y_test,y_pred),2)
```

```python
print("classification_report : ")
print(classification_report(y_test, y_pred))
```

```python
print("Confusion Matrix : ")
print(confusion_matrix(y_test, y_pred))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RF = RandomForestClassifier(n_estimators=200)
```

```python
model_RF.fit(x_train, y_train)
```

```python
y_pred1 = model_RF.predict(x_test)
```

```python
accuracy_RF = accuracy_score(y_test, y_pred1)
```

```python
print('Accuracy of training set : {:.2f}'.format(model_RF.score(x_train, y_train)))
print('Accuracy of  test set    : {:.2f}'.format(model_RF.score(x_test, y_test)))
```

```python
precision_RF = round(precision_score(y_test,y_pred1,average = 'macro'),2)
recall_RF = round(recall_score(y_test,y_pred1, average = 'macro'),2)
f1_RF = round(f1_score(y_test,y_pred1, average = 'macro'),2)
accuracy_RF = round(accuracy_score(y_test,y_pred1),2)
```

```python
print("classification_report : ")
print(classification_report(y_test, y_pred1))
```

```python
print("Confusion Matrix : ")
print(confusion_matrix(y_test, y_pred1))
```
